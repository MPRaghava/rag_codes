```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain_groq import ChatGroq
```

```python
# sample_text = """Paris is the capital and most populous city of France.
# Isaac Newton formulated the laws of motion and universal gravitation.
# Python is a programming language known for its readability and simplicity.
# """
loader = TextLoader("/content/sample_data/sample_text.txt")
docs = loader.load()
print(docs)
```

```python
# Text Splitting
text_splitter = CharacterTextSplitter(chunk_size=200, chunk_overlap=20)
texts = text_splitter.split_documents(docs)

# Embeddings and Vector store
embedding_model = HuggingFaceEmbeddings(model_name ="all-MiniLM-L6-v2")
vector_store = FAISS.from_documents(texts, embedding_model)
```

```python
!pip install langchain langchain-groq groq
```

```python
# Retrieval pipeline
retriver = vector_store.as_retriever()

llm = ChatGroq(
    api_key="key",  # Replace or use environment variable
    model_name="llama3-8b-8192"
)

rag_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriver)
```

```python
result = rag_chain.run("What is the capital of France?")
print(result)
```

```python
result = rag_chain.run("Who discovered gravity?")
print(result)
```

```python
import json
questions = [
    {
        "question":"What is the capital of France?",
        "ground_truth" : "Paris"
    },

   {
        "question": "Who discovered gravity?",
        "ground_truth": "Isaac Newton"
    }
]

results = []

for item in questions:
  answer = rag_chain.run(item["question"])
  docs = retriver.get_relevant_documents(item["question"])
  ret_texts = [doc.page_content for doc in docs]


  results.append({
      "question":item["question"],
      "answer": answer,
      "retrieved_contexts" : ret_texts,
      "ground_truth" : item["ground_truth"]
  })

  with open("rag_results.json","w") as f:
    json.dump(results,f,indent =2)
```

```python
import json
from datasets import Dataset
from ragas import evaluate
from ragas.metrics import answer_relevancy, faithfulness,context_precision,context_recall
```

```python

with open("rag_results.json" ,"r") as f:
  rag_data = json.load(f)

dataset = Dataset.from_list(rag_data)
finals = evaluate(
    dataset,
    metrics = [answer_relevancy, faithfulness,context_precision,context_recall]
)


print("📊 RAG Evaluation Metrics:")
for k,v in results.items():
  print(f"{k}: {v:.2f}")
```
